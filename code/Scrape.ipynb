```python
def parseFromUrl(book_num):
    URL = "https://www.bible.com/fr/bible/1217/GEN."+str(book_num)+".SSDC98'
    page = requests.get(URL)
    soup = BeautifulSoup(page.content, 'html.parser')
    return soup
```

```python
def getVerset(num, verset_list):
    class_value="verse v"+str(num+1)
    verset= verset_list.findAll("span", {"class": class_value})
    print(verset)
    for v in range(len(verset)):
        content=verset[v].findAll("span", {"class": "content"})[0].text.strip()
        label=verset[v].findAll("span", {"class": "label"})[0].text.strip()
        RecordtoAdd={} #initialise an empty dict 
        RecordtoAdd.update({'Verset' : label})
        RecordtoAdd.update({'Moore' : content})
        RecordtoAdd.update({'Francais' : ""})
        print(RecordtoAdd)
        rows_list.append(RecordtoAdd)

    return rows_list
```

```python
def getChapeterContent(soup):
    print("Get chapter content\n")
    soup_book=soup.findAll("div", {"class": "book bkGEN"})
    for i in range(len(soup_book)):
        soup_book_p = soup_book[i].findAll("div", {"class": "p"})
        for j in range(len(soup_book_p)):
            dt=getVerset(j, soup_book_p[j])
    
```

```python
def parseToDataFrame(numOfVerset, soup):
    rows_list=[]
    for i in range(numOfVerset):
        value="verse v"+str(i+1)
        data=soup.findAll("div", {"class": "book bkGEN"})[0].findAll("span", {"class": value})
        for j in range(len(data)): 
            label=data[j].findAll("span", {"class": "label"})
            content=data[j].findAll("span", {"class": "content"})
            RecordtoAdd={} 
            RecordtoAdd.update({'Francais' : ""})
            if len(label)>0:
                print(label[0].text.strip())
                RecordtoAdd.update({'Verset' : label[0].text.strip()})

            for c in range(len(content)):
                if c==0:
                    concat_content= content[c].text.strip()
                else:
                    concat_content+=" "+ content[c].text.strip()

                print( content[c].text.strip())
            RecordtoAdd.update({'Moore' : concat_content})
            rows_list.append(RecordtoAdd)
    return pd.DataFrame(rows_list)
```
